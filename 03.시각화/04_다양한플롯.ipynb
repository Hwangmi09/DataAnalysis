```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
```

## 1. Bar Chart
- x 데이터가 카테고리 값인 경우에 사용

```python
# plt.bar(x, y)
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

```python
# plt.barh(y) : 레이블이 긴 경우나 레이블을 세로축에 쓰는 것이 보기 편한 경우 사용 
np.random.seed(2021)

people = ['춘향전_몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4) # alpha : 투명도
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.show()
```

### 타이타닉 선실별 승객수

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head(3)
```

```python
s = titanic['class'].value_counts()
s
```

```python
df = titanic[['class', 'pclass']].groupby('class').count()
df
```

```python
y = s.values # 시리즈로 반환될 때는 내림차순으로 정렬됨
x = np.arange(len(y))
xlabel = s.index.to_list()

plt.title('타이타닉 선실별 승객수')
plt.bar(x,y)
plt.xticks(x, xlabel)
plt.xlabel('선실')
plt.ylabel('승객수')
plt.show()
```

```python
y = df.pclass.values
x = np.arange(len(y))
xlabel = df.index.to_list()

plt.title('타이타닉 선실별 승객수')
plt.bar(x,y)
plt.xticks(x, xlabel)
plt.xlabel('선실')
plt.ylabel('승객수', rotation=0, labelpad=15) # rotation=0 : 라벨을 가로로 회전, labelpad : 여백 지정
plt.show()
```

## 2. 파이 차트
- 카테고리 별 값의 상대적인 비교를 해야 할 때 사용

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]    # 합이 100이 되도록
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)    # 파이 조각이 돌출되는 크기
plt.title("Pie Chart", fontsize=15)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90) #  autopct : 파이 조각의 전체 대비 백분율, startangle : 그려지는 파이조각 시작 위치
plt.axis('equal') # 파이 차트를 그릴 때는 윈의 형태를 유지할 수 있도록 이 명령을 실행해야 한다.
plt.show()
```

### Anscombe's Quartet
- 선형회귀선 : y = 3.0 + 0.5*x

```python
import seaborn as sns
ans = sns.load_dataset('anscombe')
ans.shape
```

```python
ans.head()
```

```python
ans.groupby('dataset').describe()
```

- Dataset I에 대해서 산점도와 회귀선 그리기

```python
ds1 = ans[ans.dataset == 'I']
ds1
```

```python
plt.scatter(ds1.x, ds1.y)
plt.grid()
plt.title('Dataset I')
plt.show()
```

```python
plt.scatter(ds1.x, ds1.y)
plt.plot([2, 16], [4, 11], c='r', lw=2) # c : color, lw : line width
plt.grid()
plt.title('Dataset I')
plt.show()
```

### 4개의 데이터셋 모두에 대해서 그림

```python
fig, axes = plt.subplots(2, 2, figsize=(8,6))

for index, roman in enumerate(['I', 'II', 'III', 'IV']):
    ds = ans[ans.dataset == roman]
    ax = axes[index//2, index%2]
    ax.scatter(ds.x, ds.y)
    ax.grid()

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8,6), sharex=True, sharey=True)

for index, roman in enumerate(['I', 'II', 'III', 'IV']):
    ds = ans[ans.dataset == roman]
    ax = axes[index//2, index%2]
    ax.scatter(ds.x, ds.y)
    ax.plot([3, 20], [4.5, 13], 'r-', lw=2) # 선형회귀식 y = 0.5 + 0.3*3
    ax.set_title(f'Dataset {roman}')
    ax.grid()

fig.suptitle('Anscombe\'s Quartet', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

## 6. Box Plot

```python
iris = sns.load_dataset('iris')
iris.head()
```

### Setosa 품종의 각 feature의 박스 플롯

```python
setosa = iris[iris.species == 'setosa']
plt.figure(figsize=(8, 10))
plt.boxplot([setosa.sepal_length, setosa.sepal_width,
             setosa.petal_length, setosa.petal_width],
            labels = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])

plt.title('Setosa 품종의 각 feature별의 박스 플롯')
plt.show()
```

- 모든 품종의 feature별 박스 플롯

```python
fig, axes = plt.subplots(1, 3, figsize=(12,8), sharey=True)

for i, species in enumerate(iris.species.unique()):
    sub = iris[iris.species == species]
    ax = axes[i]
    ax.boxplot([sub.sepal_length, sub.sepal_width, sub.petal_length, sub.petal_width],
               labels=['sepal_length','sepal_width','petal_length','petal_width'])
    ax.set_title(species)

fig.suptitle("Iris 모든 품종의 피쳐별 박스 플롯", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```
