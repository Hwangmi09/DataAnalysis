```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
```

### Figure 객체

```python
np.random.seed(0)
f1 = plt.figure(figsize=(10, 2))
plt.title("figure size : (10, 2)")
plt.plot(np.random.randn(100))
plt.grid()
plt.show()
```

### Axes 객체와 subplot

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)  # Axes, subplot(2, 1, 1) : 2x1행렬, 첫번째
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
plt.grid()

ax2 = plt.subplot(2, 1, 2)  # Axes, 2x1행렬, 두번째
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
plt.grid()

plt.tight_layout() # 플롯간의 간격을 자동으로 맞춰주는 명령어
plt.show()
```

```python
np.random.seed(0)

plt.subplot(221) # 행의 갯수, 열의 갯수, 순서
plt.plot(np.random.rand(5))
plt.title("axes 1")
plt.grid()

plt.subplot(222) # (2,2,1)를 줄여서 221 라는 하나의 숫자로 표시할 수도 있다
plt.plot(np.random.rand(5))
plt.title("axes 2")
plt.grid()

plt.subplot(223) # Axes의 위치는 위에서 부터 아래로, 왼쪽에서 오른쪽으로 카운트
plt.plot(np.random.rand(5))
plt.title("axes 3")
plt.grid()

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")
plt.grid()

plt.tight_layout()
plt.show()
```

- subplots()

```python
fig, axes = plt.subplots(2, 2) # _, axes = plt.subplots(2, 2), 2x2 array

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 0].grid()
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[0, 1].grid()
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 0].grid()
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")
axes[1, 1].grid()

plt.tight_layout()
plt.show()
```

```python
np.random.seed(2021)
fig, axes = plt.subplots(2, 2, figsize=(6, 4))  # 각 그림의 사이즈

for i in range(4):
    ax = axes[i//2, i%2]
    ax.plot(np.random.rand(5))
    ax.set_title(f'axes {i+1}')
    # ax.set_xlabel('X-축'); ax.set_ylabel('Y-축')
    ax.grid()

fig.suptitle('4개의 그림', fontsize=13) # fig를 사용하는 경우
plt.tight_layout()
plt.show()
```
