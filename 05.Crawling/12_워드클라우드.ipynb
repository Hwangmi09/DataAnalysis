## 해시태그 출현 빈도

- collections 모듈

```python
from collections import Counter, OrderedDict
```

```python
s = 'A quiuk brown'
Counter(s) # 문자/단어가 몇 번 나왔는지 계산
```

```python
sl = ['A', 'quiuk', 'brown', 'brown']
Counter(sl)
```

```python
d = dict() # 순서가 없음
d['x'] = 1
d['k'] = 2
d['m'] = 3
d
```

```python
od = OrderedDict() # 순서가 있음
od['x'] = 1
od['k'] = 2
od['m'] = 3
od
```

```python
for k, v in od.items():
    print(k, v)
```

- 해시태그 빈도 집계

```python
tag_counts = Counter(tags_total)
```

```python
# Top 50
tag_counts.most_common(50) # Counter로 만들었기 때문에 사용가능
```

```python
# 데이터 정제하기
stopwords = ['#jeju', '#일상', '', '#제주', '#제주도', '#선팔', '#제주눈썹문신',
             '#서귀포눈썹문신', '#반영구', '#눈썹문신', '#제주자연눈썹', '#제주속눈썹', '#서귀포남자눈썹문신',
             '#서귀포자연눈썹', '#제주남자눈썹문신', '#서귀포속눈썹', '#서귀포반영구']
'''tag_total_refined = []
for tag in tags_total:
    if tag not in stopwords:
        tag_total_refined.append(tag)'''
tags_total = [tag for tag in tags_total if tag not in stopwords]

tag_counts = Counter(tags_total)
tag_counts.most_common(50)
```

## 워드 클라우드 만들기

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
! ls -l '/usr/share/fonts/truetype/nanum/'
```

```python
path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
wordcloud = WordCloud(font_path=path,
                      background_color='white',
                      max_words=100,
                      relative_scaling=0.3,
                      width=800, height=400
                      ).generate_from_frequencies(tag_counts)
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
